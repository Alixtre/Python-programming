### Відповіді на контрольні запитання

1. Основна відмінність між інтерпретатором і компілятором полягає в тому, що компілятор потрібно запустити тільки один раз, щоб отримати файл, який містить машинний код, а інтерпретатор потрібно запускати щоразу для запуску програми. Python – мова, що інтерпретується.

1. PyCharm - це найбільш популярна і просунута в плані функціоналу IDE для мови програмування Python, розроблена компанією JetBrains.

1. Система контроля версий - это программное обеспечение, которое помогает управлять изменениями в исходном коде с течением времени. СКВ хранит различные версии файлов, что позволяет удалять, добавлять, складывать, сворачивать, а также откатываться к указанным состояниям. GIT розподіляється, а SVN - ні. Іншими словами, кожен розробник може мати на локальній машині повну копію репозиторію. Основний плюс у тому, що якщо раптом у вас немає доступу до інтернету, зберігається можливість працювати з репозиторієм. Потім тільки один раз зробити синхронізацію і всі інші розробники отримають повну історію. 

1. При виконанні команди `git add .` всі новостворені файли додаваються у індекс.

1. При виконанні команди `git commit` зміни додаваються  у локальний Git-репозиторій.

1. За допомогою команди `git push` виконується розміщення локальної копії git-репозиторію проєкта на віддалений, що знаходиться на GitHub.
